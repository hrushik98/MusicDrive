```python
def mp3_downloader(link):

  import re

  text = f"{link}"

  # Define a regular expression pattern to match YouTube URLs
  youtube_url_pattern = r'https://(?:www\.)?youtube\.com/watch\?v=([\w-]+)'

  # Use re.search to find the first match in the text
  match = re.search(youtube_url_pattern, text)

  # Check if a match was found
  if match:
      # Extract the YouTube video ID from the match
      youtube_video_id = match.group(1)

      # Construct the full YouTube URL using the extracted video ID
      youtube_full_url = f'https://youtube.com/watch?v={youtube_video_id}'
  else:
      print("No YouTube URL found in the text.")
  # importing packages
  from pytube import YouTube
  import os

  # url input from user
  yt = YouTube(str(youtube_full_url) )

  # extract only audio
  video = yt.streams.filter(only_audio=True).first()

  # check for destination to save file

  # destination = str(input(">> ")) or '.'
  destination = "."

  # download the file
  out_file = video.download(output_path=destination)

  # save the file
  base, ext = os.path.splitext(out_file)
  new_file = base + '.mp3'
  os.rename(out_file, new_file)

  # result of success
  print(yt.title + " has been successfully downloaded.")
```
